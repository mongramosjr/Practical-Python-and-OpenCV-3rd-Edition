```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

image = cv2.imread("trex3.png")

print("width: {} pixels".format(image.shape[1]))
print("height: {} pixels".format(image.shape[0]))
print("channels: {}".format(image.shape[2]))
```

```python
#Matplot lib expects img in RGB format but OpenCV provides it in BGR.
#img[:,:,::-1] is numpy slicing, which means that we select all the elements on the first axis, 
#all the elements on the seconds, and then the reverse of the elements on the third axis.
image_rgb = image[:,:,::-1]
plt.imshow(image_rgb)
```

```python
image_rgb2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```
