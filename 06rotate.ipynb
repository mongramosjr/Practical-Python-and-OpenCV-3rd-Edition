```python
import numpy as np
import imutils
import cv2
from matplotlib import pyplot as plt

image = cv2.imread("t-rex.jpg")
```

```python
# print original image
image_rgb = image[:,:,::-1]
plt.imshow(image_rgb)
```

```python
(h,w) = image.shape[:2]
center = (w // 2, h // 2)

M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(image, M, (w,h))

#cv2.imshow("Rotated by 45 degress", rotated)
image_rgb = rotated[:,:,::-1]
plt.imshow(image_rgb)
```

```python
M = cv2.getRotationMatrix2D(center, -90, 1.0)
rotated = cv2.warpAffine(image, M, (w,h))
#cv2.imshow("Rotated by -90 degress", rotated)
image_rgb = rotated[:,:,::-1]
plt.imshow(image_rgb)
```

```python
rotated = imutils.rotate(image, 180)
#cv2.imshow("Rotated by 180 degress", rotated)
image_rgb = rotated[:,:,::-1]
plt.imshow(image_rgb)
```
